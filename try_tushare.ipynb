```python
from datetime import datetime, timedelta
def strategy(ts_code, current_date, obs_days=5, threshould=3):
    
    def start_date(end_date, days):
        # 将 end_date 转换为 datetime 对象
        end_date_obj = datetime.strptime(end_date, '%Y%m%d')
        # 计算开始日期
        start_date_obj = end_date_obj - timedelta(days=days)
        # 将开始日期格式化为字符串并返回
        return start_date_obj.strftime('%Y%m%d')
    
    df = pro.daily(ts_code=ts_code, start_date=start_date(current_date, obs_days), end_date=current_date)
    while df.shape[0] < obs_days + 1:
        df = pro.daily(ts_code=ts_code, start_date=start_date(current_date, obs_days + df.shape[0]), end_date=current_date)

    df = df.sort_values('trade_date', ascending=True)
    df['increase_stop'] = round(df['pre_close'] * 1.1, 2)
    df['vol_change'] = df['vol'].diff()
    # 放缩
    df.loc[df['vol_change'] > 0, '放缩'] = '放'
    df.loc[df['vol_change'] < 0, '放缩'] = '缩'
    # 涨跌
    df.loc[df['change'] > 0, '涨跌'] = '涨'
    df.loc[df['change'] < 0, '涨跌'] = '跌'
    # 是否涨停
    df.loc[df['close'] == df['increase_stop'], '是否涨停'] = '是'
    df.loc[df['close'] != df['increase_stop'], '是否涨停'] = '否'
    # flag
    df['flag'] = 0
    df.loc[(df['放缩'] == '放') & (df['涨跌'] == '涨'), 'flag'] = 1
    df.loc[(df['放缩'] == '缩') & (df['涨跌'] == '跌'), 'flag'] = 1
    df.loc[(df['放缩'] == '缩') & (df['涨跌'] == '涨') & (df['是否涨停'] == '是'), 'flag'] = 1
    # name
    tmp = stock_info[['name', 'ts_code']].merge(df, on='ts_code', how='inner')
    res = tmp.iloc[-threshould:, ]
    ans = 0
    if res['flag'].sum(axis=0) == threshould:
        ans = 1
    res_dict = {
        'detail': tmp.iloc[-obs_days:, ],
        'ans': ans
    }
    return res_dict
```

```python
w = pd.ExcelWriter('./20241104.xlsx')
ls = []
for ts_code in ts_code_ls:
    d = strategy(ts_code, current_date='20241104')
    sheet_name = selected_stock.loc[selected_stock['代码']==ts_code, '名称'].values[0]
    d['detail'].to_excel(w, sheet_name=sheet_name, index=False)
    if d['ans'] > 0:
        ls.append(sheet_name)
print(ls)
w.close()
```
